## Theoretical Part

Please answer the following questions. 

#### What are the assumptions of a linear model (or any other type of model)?

Principal assumptions:
- linearity: the relationship between X and the mean of Y is linear
- homoscedasticity: the variance of residual is the same for any value of X
- independence: observations are independent

#### What’s the difference between K Nearest Neighbor and K-means Clustering?

KNN and K-Means Clustering both represent distance-based algorithms but KNN is used in supervised learning for classification while K-means clustering is mainly used for clustering in unsupervised learning.

- KNN: give new data points accordingly to the k number or the closest data points.
- K-Means Clustering: groups data into k number of clusters.

#### How do you address overfitting?

There is several techniques to address overfitting:
- Add more data if available. In the case that we cannot find more data, there are techniques we can apply, called Data Augmentation to artificially add sufficiently similar data points to our dataset. 
- Reduce the number of features by excluding those features less related to the outcome, and therefore reducing the amount of noise to be modeled the dataset. 
- In extreme cases, we might want to use a dimensionality reduction technique, such as Principal Component Analysis (PCA), to reduce the size of our data. 
- Apply regularization techniques to the model

#### Explain Naive Bayes algorithms.

Naive Bayes algorithms are classification algorithms based on Bayes' Theorem. They all of them share a common principle: every pair of features being classified is independent.

#### When do you use an AUC-ROC score? What kind of information can you gather from it?

AUC-ROC score is a classification model’s performance metric.
It gives you the rate of false positives and false negative.

#### What is cross validation?

Cross-validation is used to prevent overfitting. The idea is to use the training data to generate multiple mini train-test splits that you will use to tune the model.

#### What are confounding variables?

A confounding variable is an outside influence that changes the effect of a dependent and independent variable.

#### If an important metric for our company stopped appearing in our data source, how would you investigate the causes?

Try and determine if the metric was deleted on purpose by contacting the other persons working on the data source

### Preprocessing

```python
# Dropping Id column
housing.drop('Id', axis='columns', inplace=True)
```

```python
# Looking into null values
for column in housing.columns:
    if housing[column].isnull().any():
        print('{0} has {1} null values'.format(column, housing[column].isnull().sum()))
```

```python
# Dropping columns with more than 10% of the total values missing
to_drop = [col for col in housing.columns if housing[col].isnull().sum() > housing.shape[0] / 10]
print(to_drop)

housing.drop(to_drop, axis='columns', inplace=True)
```

```python
# Filling remaining null values with 0
housing = housing.fillna(0)
```

```python
# Object columns
object_cols = [col for col in housing if housing[col].dtypes == 'object']
print("Number of object columns: ", len(object_cols))
print(object_cols)
```

```python
# Label: 'Neighborhood', 'Heating', 'CentralAir', 'KitchenQual'
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

housing['Neighborhood_label'] = le.fit_transform(housing['Neighborhood'])
housing['Heating_label'] = le.fit_transform(housing['Heating'])
housing['CentralAir_label'] = le.fit_transform(housing['CentralAir'])
housing['KitchenQual_label'] = le.fit_transform(housing['KitchenQual'])
```

```python
housing.drop(object_cols, axis='columns', inplace=True)
```

```python
# Correlation matrix
corr = housing.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize = (11, 9))

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype = np.bool))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap = True)

# Draw the heatmap with the mask
sns.heatmap(corr, cmap = cmap, mask = mask, center = 0, square = True)

plt.show()
```

```python
#High correlation with target
corr_target = abs(corr["SalePrice"])
corr_target.sort_values(ascending = False)

# No correlations of 0.9 or more
```

### Modeling

```python
# Selecting features for modeling
X = housing.drop('SalePrice', axis = 'columns')
y = housing['SalePrice']
```

```python
# Standarize features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.2)
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

lr = LinearRegression()

# Fit the model to the training data
lr.fit(X_train, y_train)

# Evaluate the model with R-Squared score on X_train and X_test data 
y_train_pred = lr.predict(X_train)
r2_train = r2_score(y_train, y_train_pred)

y_test_pred = lr.predict(X_test)
r2_test = r2_score(y_test, y_test_pred)

# Compare the value to evaluate the model
print("R Squared score:", round(r2_train, 4))
print("R Squared score:", round(r2_test, 4))

print(f"Linear Regression Test Accuracy: {round(lr.score(X_test, y_test) * 100, 2)}%")

# The model might be a little overfitted
```

```python
# Increase training data to 90%
X_train09, X_test09, y_train09, y_test09 = train_test_split(X, y, test_size = 0.1)
```

```python
lr09 = LinearRegression()

# Fit the model to the training data
lr09.fit(X_train09, y_train09)

# Evaluate the model with R-Squared score on X_train and X_test data 
y_train_pred09 = lr09.predict(X_train09)
r2_train09 = r2_score(y_train09, y_train_pred09)

y_test_pred09 = lr09.predict(X_test09)
r2_test09 = r2_score(y_test09, y_test_pred09)

# Compare the value to evaluate the model
print("R Squared score:", round(r2_train09, 4))
print("R Squared score:", round(r2_test09, 4))

print(f"Linear Regression Test Accuracy: {round(lr09.score(X_test09, y_test09)*100  , 2)}%")
```

#### Decision Tree

```python
# Create DecisionTree regression
regr = DecisionTreeRegressor(random_state = 29)

# Fit the model
regr.fit(X_train, y_train)

# Predict observation's value  
acc = regr.score(X_test, y_test) * 100
print(f"Decision Tree Test Accuracy {round(acc, 2)}%")
```

```python
# Create DecisionTree regression
regr09 = DecisionTreeRegressor(random_state = 29)

# Fit the model
regr09.fit(X_train09, y_train09)

# Predict observation's value  
acc09 = regr09.score(X_test09, y_test09) * 100
print(f"Decision Tree Test Accuracy {round(acc09, 2)}%")

# Decision Tree works better with less training data
```

The Linear Regression with more training data is the model that performed the best, now let's try with the outliers removed.

```python
# Without outliers
Q1 = housing['SalePrice'].quantile(0.25)
Q3 = housing['SalePrice'].quantile(0.75)
IQR = Q3 - Q1

no_outliers = housing[~((housing['SalePrice'] < (Q1 - 1.5 * IQR)) | 
                                    (housing['SalePrice'] > (Q3 + 1.5 * IQR)))]
```

```python
# Selecting features for modeling
X = no_outliers.drop('SalePrice', axis = 'columns')
y = no_outliers['SalePrice']
```

```python
# Standarize features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size = 0.2)
```

```python
lr = LinearRegression()

# Fit the model to the training data
lr.fit(X_train, y_train)

# Evaluate the model with R-Squared score on X_train and X_test data 
y_train_pred = lr.predict(X_train)
r2_train = r2_score(y_train, y_train_pred)

y_test_pred = lr.predict(X_test)
r2_test = r2_score(y_test, y_test_pred)

# Compare the value to evaluate the model
print("R Squared score:", round(r2_train, 4))
print("R Squared score:", round(r2_test, 4))

print(f"Linear Regression Test Accuracy: {round(lr.score(X_test, y_test) * 100, 2)}%")
```

```python
# Increase training data to 90%
X_train09, X_test09, y_train09, y_test09 = train_test_split(X, y, test_size = 0.1)
```

```python
lr09 = LinearRegression()

# Fit the model to the training data
lr09.fit(X_train09, y_train09)

# Evaluate the model with R-Squared score on X_train and X_test data 
y_train_pred09 = lr09.predict(X_train09)
r2_train09 = r2_score(y_train09, y_train_pred09)

y_test_pred09 = lr09.predict(X_test09)
r2_test09 = r2_score(y_test09, y_test_pred09)

# Compare the value to evaluate the model
print("R Squared score:", round(r2_train09, 4))
print("R Squared score:", round(r2_test09, 4))

print(f"Linear Regression Test Accuracy: {round(lr09.score(X_test09, y_test09)*100  , 2)}%")
```
